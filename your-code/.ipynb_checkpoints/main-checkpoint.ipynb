### Challenge 1: Fork Languages

```python
url = 'https://api.github.com/repos/ironhack-datalabs/bcn-feb-2019'

response = requests.get(url)
response
```

#### 1. Obtain the full list of forks created from the main lab repo via Github API

From the web https://api.github.com/ we can obtain the information available in the APIs. For this question, we are interested in forks_url so we will get it from https://api.github.com/repos/ironhack-datalabs/bcn-feb-2019/forks. We can see that forks_count = 7, so our list of forks should match this figure.

```python
response.json()
```

```python
forks = requests.get('https://api.github.com/repos/ironhack-datalabs/bcn-feb-2019/forks').json()
forks
```

With the code above we have obtained the list of forks, but it is difficult to read. Let's store the info on a DataFrame:

```python
fork_df = pd.json_normalize(forks)
pd.set_option('display.max_columns', None)
fork_df
```

#### 2. Loop the JSON response to find out the language attribute of each fork. Use an array to store the language attributes of each fork.

```python


lan_urls = fork_df['languages_url']
lan_keys = []

for page in range(0, len(lan_urls)):
    print(lan_urls[page])
    print(requests.get(lan_urls[page]).json().keys())
    if list(requests.get(lan_urls[page]).json().keys()) in lan_keys:
        continue
    else:
        lan_keys.append(list(requests.get(lan_urls[page]).json().keys()))
```

```python
print(lan_keys)
```
